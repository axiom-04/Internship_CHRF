```python
df.drop(columns=['Zila','Address','UniqueID','month','AdmDate','Age (month)','Case Definition','Outcome','Discharge Date','Hospital Duration','organism'],inplace=True)
```

```python
by_comp = df.groupby("year")
df1=by_comp.sum()
df1.reset_index()
```
